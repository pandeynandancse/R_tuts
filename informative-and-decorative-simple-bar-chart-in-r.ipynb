# 1. Problem Statement -
Suppose you have a data set in which there are a lot of features and one (or more) of them is (are) a categorical Feature. **My question is :** How such feaature(s) is(are) analyzed visually one by one? 

# 2. Solution of Problem -
Such feature(s) is(are) analyzed visually one by one with the help of Simple Bar Chart.

# 3. Simple Bar Chart -
Simple Bar Chart provide visual representation of a categorical feature at an instance. It generally have -
1. Categories on X-axis
2. Numbers(frequency or relative frequency) of the respective category on Y-axis

Such type of Simple Bar Chart is also known as **Vertical Bar Chart** or **Column Chart**.

# 4. When it is used - 
Use this chart type when -
1. You have to compare numbers(frequencies or relative frequencies) across a few categories of a categorical feature/variable.
2. The order of categories is not important.

# 5. Parts of a Simple Bar Chart -
Before plotting a chart, one must understand the different parts of a chart. Mainly, there are six parts of a Simple Bar Chart -
1. Main Title
2. Sub Title
3. Axes
4. Axis Titles
5. Axis labels
6. Data and Data Labels (Data Panel)

**1. Steps to create a Simple Bar Chart -**
* Step - (i) : Create a frequency table (**"table(column_name)" function is used**) for Survived and assign this table in R-object "a"
* Step - (ii): Create Simple Bar Chart by **barplot(a)** function, passing R-object already created in step - (i) as its argument and assign this plot in R-object  "x"  for further use.

```r
# Step - (i)

# Creating frequency table for Survived and assigning this table in R-object "a" 
a <- table(data$Survived)

# Print object a
a
```

**The above output shows the frequency table for "Survived". It means that in "Survived" column of "data" 0 ocuurs  549 times and 1 occurs 342 times.**

```r
# Step - (ii)

# Creating Simple bar chart and assign it in R-object "x" for further use
x <- barplot(a)
```

 Well, Simple Bar Chart is now ready but it is incomplete. **Why ?**

**Let us think, you are the person who know about the data set and if you create such type of chart, you may be able to know what this chart is explaining. But if you show the same chart to a person who does not know about your data set, then he can not understand your chart. And that's why there is a need to make it more informative.**

**2. Steps to make the above obtained Simple Bar Chart more Informative -**
* Step-(i)   : Adjust Limit of Y-axis  ---> use **ylim = c(lower_limit,upper_limit)** argument in **barplot()** function.
* Step-(ii)  : Rename Labels of X-axis ---> use **names.arg = c("label_for_0" , "label_for_1")** argument in **barplot()** function.
* Step-(iii) : Add all titles (Main title, Sub title , Axis titles) ---> use **title()** function after **barplot()** function.

**Arguments for title() function -**
1. Add Main Title ---> use **main = "Main_title"** argument in **title()** function.
2. Add Sub Title  ---> use **sub = "Sub_title"** argument in **title()** function.
3. Add X-axis Title ---> use **xlab = "X-axis_title"** argument in **title()** function.
4. Add Y-axis Title ---> use **ylab = "Y-axis_title"** argument in **title()** function.


* Step-(iv)  : Add data labels ---> use **text()** function after **barplot()** function.

**Arguments for text(d1 , d2 , d3) function -**
1. d1 ---> Represents the R-object containing the barplot. (In my case it is : x)
2. d2 ---> Represents the position of Labels, i.e., where we want to write the lablels in plot. In my case it is : a + 16 , where a : R-object containing frequency table of the underlying categorical feature and i have added 16 to a which gives a suitable position for labels for all categories.

> **Note :** The value : 16 varies from plot to plot and is obtained by hit and trial method, i.e., First try a + 1 and plot the graph. If it is well, then good , Otherwise Increase it untill you get a suitable place for labels.

3. d3 ---> Represents labels what you want to write as a character. Use **labels = as.character(a)** in place of d3, where a : R-object containing frequency table. 

```r
# Step-(i)

# Increasing limit of y-axis
x <- barplot(a , ylim = c(0,600))
```

```r
# Step-(ii)

# Renaming X-axis Labels
x <- barplot(a , ylim = c(0,600) , names.arg = c("Not Survived","Survived"))
```

```r
# Step-(iii)

x <- barplot(a , ylim = c(0,600) , names.arg = c("Not Survived","Survived"))

# Add all type of titles
title(main = "Distribution of Survived" , sub = "Fig : Bar Chart" , xlab = "Info about Survival of a Passenger ----->" , ylab = "Frequency ----->" )      
```

```r
# Step-(iv)

x <- barplot(a , ylim = c(0,600) , names.arg = c("Not Survived","Survived"))

# Add all type of titles
title(main = "Distribution of Survived" , sub = "Fig : Bar Chart" , xlab = "Info about Survival of a Passenger ----->" , ylab = "Frequency ----->" )
      
# Add data labels
text(x , a+16, labels = as.character(a))
```

**Why decorative chart ?**

Generally, People love with colours. A decorative chart as well as informative attracts the attention of people more than an informative chart. Hence, there is also a need of Decorative Chart.

**3. Steps to make the above obtained informative Simple Bar Chart as a Decorative chart -**
* **Step-(i)**   : Rotate Y-axis labels for a better view ---> use **las = 1** as argument in **barplot()** function.

* **Step-(ii)** : Decorate Rectangular Bars ---> 3-more arguments are used in **barplot()** function.

**Argument Details -**
1. **col** ---> To colour Rectangular Bars
2. **border** ---> To colour border of Rectangular Bars
3. **density** ---> To adjust the density of lines in each bars


* **Step-(iii)** : Decorate Axis Labels ---> 4-more arguments are used in **barplot()** function.

**Argument Details -**
1. **cex.axis** ---> To adjust size of Y-axis Labels
2. **cex.names** ---> To adjust size of X-axis Labels
3. **font.axis** ---> To change font style of Axes (font = 1 for Normal, font = 2 for Bold, font = 3 for Italic, font = 4 for Bold and Italic)
4. **col.axis** ---> To change the colour of Axis Labels


* **Step-(iv)**  : Decorate Titles ---> 9-more arguments are used in **title()** function.

**Argument Details -**
1. **cex.main** ---> To adjust size of Main title
2. **cex.sub** ---> To adjust size of Sub title
3. **cex.lab** ---> To adjust size of Axis titles
4. **font.main** ---> To change font style of Main title
5. **font.sub** ---> To change font style of Sub title
6. **font.lab** ---> To change font style of Axis titles
7. **col.main** ---> To change the colour of Main title
8. **col.sub** ---> To change the colour of Sub title
9. **col.lab** ---> To change the colour of Axis titles


* **Step-(v)**   : Decorate Data Labels ---> 3-more arguments are used in **text()** function.

**Argument Details -**
1. **cex** ---> To adjust size of Data Labels
2. **font** ---> To change font style of Data Labels
3. **col** ---> To change the colour of Data Labels

```r
# Step-(i)

# Rotate Y-axis labels
x <- barplot(a , ylim = c(0,600) , names.arg = c("Not Survived","Survived") , las = 1)

title(main = "Distribution of Survived" , sub = "Fig : Bar Chart" , xlab = "Info about Survival of a Passenger ----->" , ylab = "Frequency ----->" )    

text(x , a+16, labels = as.character(a))
```

# 8. Some important points -
1. Till now, I have made Simple Bar Chart using frequency. **You may also use Relative Frequency**. In this case, only one difference is in creating the R-object containing frequency table. **Now, you have to create a R-object containing relative frequency table and this is achieved as follows -**

```r
# Creating Relative frequency table and assigning it in a R-object "a"
a <- table(data$Survived)/length(data$Survived)

# Print the object a
a
```

**The above output shows the Relative Frequency Table for Survived feature.**

**All other steps remains same as in the above case.**

2. X-axis and Y-axis can be interchanged, i.e., we may represent different categories on Y-axis and the numbers of the respective category on X-axis. Such chart is known as **Horizontal Simple Bar Chart**. To plot it, just pass one more argument **horiz = TRUE** in **barplot()** function and make some changes accordingly, such as X-axis and Y-axis title, position of data labels etc. (Do it yourself)
3. Horizontal Simple Bar Chart is prefered over Vertical Simple Bar Chart when there are a large number of categories to be compared.
4. Height(Length) of bars is proportional to the numbers (frequency/relative frequency) that they represent.
5. The width of bars are arbitrary.
6. Bar charts are usually scaled so that all the data can fit properly on the chart. (This concept is used while you are working on a graph-paper)
7. The values of arguments like cex, cex.main, cex.sub, cex.lab, cex.axis, cex.names, density are chosen by hit and trial method. (No need to remember such values)
8. The values for **font** and **col** depends on your choice. You may use **Hexcode of colours** also to choose your favourite colour. Example : **Hexcode for deep pink - #FF1493**

> Link to Hexcode of colours : https://www.rapidtables.com/web/color/RGB_Color.html
